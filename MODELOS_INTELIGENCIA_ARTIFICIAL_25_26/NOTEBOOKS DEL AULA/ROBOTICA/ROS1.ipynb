## 3.3. Detectando color


```python
import cv2, numpy as np

cam = cv2.VideoCapture(0)

while True:
    ret, frame = cam.read()
    if not ret:
        break

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Detectar rojo
    lower = np.array([0,120,70])
    upper = np.array([10,255,255])

    mask = cv2.inRange(hsv, lower, upper)

    cv2.imshow("Original", frame)
    cv2.imshow("Mascara rojo", mask)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```

# ü§ñ Unidad 4 ‚Äî Integraci√≥n: Visi√≥n + Movimiento

## 4.1. Idea clave

---
## 4.2. Ejemplo: el brazo saluda cuando detecta rojo


```python
import cv2, numpy as np, time
from dofbot_control import ServoController

c = ServoController()
BASE = 1
CODO = 3

def saludo():
    for i in range(3):
        c.move_servo(CODO, 60)
        time.sleep(0.3)
        c.move_servo(CODO, 120)
        time.sleep(0.3)
    c.move_servo(CODO, 90)

cam = cv2.VideoCapture(0)

while True:
    ret, frame = cam.read()
    if not ret:
        break

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower = np.array([0,120,70])
    upper = np.array([10,255,255])
    mask = cv2.inRange(hsv, lower, upper)

    red_pixels = cv2.countNonZero(mask)

    if red_pixels > 5000:
        saludo()
        time.sleep(2)

    cv2.imshow("Camara", frame)
    cv2.imshow("Mascara", mask)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
