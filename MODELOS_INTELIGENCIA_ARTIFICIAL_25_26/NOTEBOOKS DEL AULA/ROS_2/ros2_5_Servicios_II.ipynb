# ROS 2 - Apuntes: Servicios.

Los servicios son otro método de comunicación para los nodos del gráfico ROS.    

Los servicios se basan en un modelo de llamada y respuesta, frente al modelo editor-suscriptor de los temas.    

Mientras que los temas permiten a los nodos suscribirse a flujos de datos y obtener actualizaciones continuas, los servicios *solo proporcionan datos cuando son llamados* específicamente por un cliente.

### Tareas

1. Setup   
Levantamos los dos nodos de 'turtlesim', /tutlesim and /teleop_turtle.

En un terminal:
```bash   
ros2 run turtlesim turtlesim_node
```

En otro terminal:

```bash   
ros2 run turtlesim turtle_teleop_key
```
2. Lista de servicios   

En un nuevo terminal, ejecutamos el listado de servicios:   
```bash
ros2 service list
```
Comprobamos que ambos nodos tienen los mismos seis servicios con *parámetros* en sus nombres. Casi todos los nodos de ROS 2 tienen estos servicios de infraestructura en los que se basan los parámetros. 

3. Tipos de servicios   

Los servicios tienen **tipos** que describen cómo se estructuran los datos de solicitud y respuesta de un servicio.     

Los tipos de servicio se definen de manera similar a los tipos de topics, excepto que los tipos de servicio tienen dos partes: un *mensaje para la solicitud* y *otro para la respuesta*.

Para averiguar el tipo de un servicio, se usa el comando:   

```bash   
ros2 service type <service_name>   
```    
Si, aprovechando `turtlesim`, ejecutamos lo siguiente:

```bash   
ros2 service type /clear
std_srvs/srv/Empty
```
El tipo **Empty** significa que la llamada al servicio no envía datos al realizar una solicitud y no recibe datos al recibir una respuesta.
